```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet(r"../donnees_exemple/log_export.parquet")

# suppresion des lignes o√π interface_in est nul
df = df.dropna(subset=['regle'])

print(df.shape)

df.head()
```

```python
import matplotlib.pyplot as plt

# # stats et graphs sur portsrc et portdst
# print(df['portsrc'].describe())
# print(df['portdst'].describe())

df['portsrc'].plot.hist(bins=100, alpha=0.5)
plt.show()

# ports les plus utilis√©s
print(df['portsrc'].value_counts().head(10))
print(df['portdst'].value_counts().head(10))
```

```python
# graph proto, regle et action, affichage des pourcentages dans les pie charts

df['proto'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()

df['regle'].value_counts().plot(kind='bar')
plt.show()

df['action'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```
